```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt


# Load your DataFrame
data = pd.read_csv('WK_1-6.csv')  # Replace 'your_data.csv' with the actual file path
# 'TEAM_PTS','PROJ_PTS','PASS_COMP','PASS_YDS','PASS_TD','INT','CAR','RUSH_YDS','RUSH_TD','REC','REC_YARDS','REC_TD','TARGETS','2PC','FUML','MISC_TD',
data = data.drop(columns=['TEAM_PTS','PROJ_PTS','PASS_COMP','PASS_YDS','PASS_TD','INT','CAR','RUSH_YDS','RUSH_TD','REC','REC_YARDS','REC_TD','TARGETS','2PC','FUML','MISC_TD',"DnP"])

label_encoders = {}
categorical_columns = ['POS', 'TEAM', 'OPP', 'NAME', '@HOME']
for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column])


# Compute the correlation matrix
correlation_matrix = data.corr()


# Create a heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
